# Jupyter Notebook Demo

Glyph renders `.ipynb` files directly. **Markdown cells** support the full markdown feature set, including inline math like $e^{i\pi} + 1 = 0$ and block math:

$$\int_0^\infty e^{-x^2}\,dx = \frac{\sqrt{\pi}}{2}$$

...and Mermaid diagrams:

```mermaid
flowchart LR
  A[Read .ipynb] --> B{Cell type}
  B -->|markdown| C[MarkdownContent]
  B -->|code| D[Highlighted source]
  D --> E[Outputs]
```
